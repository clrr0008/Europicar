#2 . Unimos los archivos csv:
Unimos los 4 archivos csv en uno para trabajar comodamente:

Cada uno de los archivos contiene una cabecera y solo queremos la del primer archivo.

```python
# Función para unir los archivos csv en uno solo.

def concatenar_lista_archivos(lista_archivos):
  global df
  df = pd.read_csv(lista_archivos[0])

  for i in range (1, len(lista_archivos)):
    if i <= len(lista_archivos):
      df1 = pd.read_csv(lista_archivos[i])
      df = pd.concat([df,df1]) # concatenamos un nuevo archivo al conjunto ya concatenado y reiteramos hasta completarlo.
  return df


archivos = ["CSV_Scraping_autohero.csv", "CSV_Scraping_autohero2.csv", "CSV_Scraping_autohero3.csv", "CSV_Scraping_autohero4.csv"]
concatenar_lista_archivos(archivos)
```

Cambiamos diferentes valores de las columnas a sus valores correctos:

```python
# Cambiamos los diferentes valores de las columnas a sus valores correctos, limpios para analizar:


df["Marca"] = df["Marca"].map(lambda x: x.replace(",","").strip().split(" ")[0])

df["Modelo"] = df["Modelo"].map(lambda x: x.strip().replace(",", "").split(" ")[1:])
df["Modelo"] = df["Modelo"].map(lambda x: " ".join(x))

df["Version"].map(lambda x: x.strip())

# 'Precio' tiene como ejemplo de valores esto: '28.399\xa0€'
# Limpiamos todo esto para quedarnos solo con "28399"
df["Precio"] = df["Precio"].map(lambda x: x.replace("\xa0€","").replace(".","").strip())

df["Combustible"] = df["Combustible"].map(lambda x: x.strip())

#Columna "Año" es algo más enrevesado, lo hacemos más abajo, aparte.

#Columna 'Kms' tiene como ejemplo de valor: '61.395 km', vamos a quedarnos con: '61395'
df["Kms"] = df["Kms"].apply(lambda x: x.strip().split(" ")[0].replace('.',''))

df["Potencia"] = df["Potencia"].apply(lambda x: x.strip().split(" ")[0])

# Columna "Puertas" vamos a unir los valores "2" y "4" con los de "3" y "5", respectivamente.
# Los valores únicos de "Puertas" son ['5', '3', '2', '4', '5.0', '4.0', '2.0', '3.0'].
# Antes de cambiar valores, lo pasamos a cadena todo (aunque ya está como objeto) porque me da error debido a los .0 por alguna razón.
df["Puertas"] = df["Puertas"].astype(str)
df['Puertas'] = df['Puertas'].apply(lambda x: x.replace('.0','')) # quitamos los '.0' de algunos valores.

df["Puertas"].loc[df[df["Puertas"] == '2'].index] = '3'
df["Puertas"].loc[df[df["Puertas"] == '4'].index] = '5'

#Columna "Cambio" tiene los valores únicos:'Cambio tipo automático','Cambio tipo manual','Semi-automático','Doble embrague'.
#Vamos a convertirlos a: "automático, manual, semi-automático y doble embrague"
def convertir_cambio(cambio):
  if cambio == "Cambio tipo automático":
    cambio = "Automático"
  elif cambio == "Cambio tipo manual":
    cambio = "Manual"
  elif cambio == "Semi-automático":
    cambio = "Semi-automático"
  elif cambio == "Doble embrague":
    cambio = "Doble embrague"
  else: print("Algo salió mal!, debe de haber algún espacio en blanco entre las cadenas")
  return cambio
df["Cambio"] = df["Cambio"].apply(lambda x: x.strip())
df["Cambio"] = df["Cambio"].apply(lambda x: convertir_cambio(x))
df["Cambio"]

#Columna "Color" está correcta aunque vamos a limpiarle por si tuviera algún espacio en blanco escondido
df["Color"] = df["Color"].apply(lambda x: x.strip())

#Columna "Distribuidor" está correcta puesto que la creamos nosotros mismos.
```

### Columna "Año":

```python
df["Año"].head(2)
```

```python
# Convertimos en string y nos quedamos con el año

df["Año"] = df["Año"].astype(str)

df["Año"] = df["Año"].apply(lambda x: x.split(".")[-1])
df["Año"].head()
```

```python
# Le añadimos un "0" a los años que al pasarlos a cadena, se les quitó el cero de la derecha.
def añadir_cero(string):
  if len(string) < 4:
    string = string + "0"
  return string

df["Año"] = df["Año"].apply(lambda x: añadir_cero(x))
df["Año"].head()
```

## Transformación columnas, tipo de datos:

Transformación de datos a valores correctos.

```python
df.dtypes
```

Todas son objetos, vamos a transformar las que no deben serlo a sus tipos correspondientes:

1. "Precio"(object) -------> int64.
2. "Año"(object) -----------> datetime[ns]
3. "Kms"(object) ----------> Float64.
4. "Potencia"(object)----> Int64

```python
df.dtypes
```

```python
# Convertimos a la vez todas las columnas a su tipo correspondiente:
# Para la columna "Año" lo hacemos diferente, más abajo.
# Para convertir el precio a un int(), nos encontramos con que hay precios que contienen un "." como separador de miles.
# Limpiamos previamente este "." y despues convertimos:

df['Precio'] = df['Precio'].apply(lambda x: x.replace(".", ""))

df = df.astype({"Precio":"int64", "Potencia":"Int64", "Kms":"Float64"})
df["Año"] = pd.to_datetime(df["Año"], format='%Y')
```
