# Agrupaciones para grÃ¡fico dispersion bÃºrbujas POWER BI


```python
agrupacion_csv = df_csv.groupby('Marca').agg({'Marca':'count','Precio':'mean'})
agrupacion_api = df_api.groupby('Marca').agg({'Marca':'count','Precio':'mean'})

agrupacion_csv
# agrupacion_csv.merge(agrupacion_api, on = 'Marca', how = 'outer', suffixes = ('_csv','_api')).fillna(0)
```

```python
agrupacion_csv = agrupacion_csv.rename(columns={'Marca':'Marcas'})
agrupacion_api = agrupacion_api.rename(columns={'Marca':'Marcas'})
```

```python
# agrupacion_csv.merge(agrupacion_api, on = 'Marca', how = 'outer', suffixes = ('_csv','_api')).fillna(0)
```

```python
agrupacion_csv.merge(agrupacion_api, on = 'Marca', how = 'outer', suffixes = ('_csv','_api')).fillna(0).reset_index().to_csv('agrupacion_csv_y_api.csv',
                                                                                                        index=False, sep=';')
```

```python
df_merge = pd.read_csv('agrupacion_csv_y_api.csv', sep=';')
```

```python
df_merge['Precio_api'] = df_merge['Precio_api'].apply(lambda x: x*12*4)
```

```python
# convertir a entero los precios para que no salgan millones de decimales:

for col in df_merge.columns:
  if col == 'Precio_csv' or col == 'Precio_api' or col == 'Marcas_csv' or col == 'Marcas_api':
    df_merge[col] = df_merge[col].astype(int)
```
